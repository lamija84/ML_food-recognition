```python
# ======================================================================
# 0) IMPORTING LIBRARIES AND CHECKING GPU
# Description: Load required libraries and print TensorFlow/GPU info.
# ======================================================================

import os, zipfile, shutil
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("TensorFlow:", tf.__version__)
print("GPUs:", tf.config.list_physical_devices("GPU"))

from google.colab import drive
drive.mount('/content/drive')


# ======================================================================
# 1) UNZIPPING THE DATASET AND EXPLORING DIRECTORIES
# Description: Extract the dataset and determine correct root directory.
# ======================================================================

zip_path = "/content/drive/MyDrive/datasets/food-11.zip"
extract_root = "/content/food-11"

os.makedirs(extract_root, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_root)

print("Contents of /content/food-11:", os.listdir(extract_root))

inner_dir = os.path.join(extract_root, "food-11")
DATA_DIR = inner_dir if os.path.isdir(inner_dir) else extract_root

print("DATA_DIR:", DATA_DIR)
print("DATA_DIR contents:", os.listdir(DATA_DIR))


# ======================================================================
# 2) FIXING DIRECTORY STRUCTURE
# Description: Ensures every split contains class folders (0–10).
# Moves images into their corresponding class directories.
# ======================================================================

splits = ["training", "validation", "evaluation"]

for split in splits:
    split_dir = os.path.join(DATA_DIR, split)
    if not os.path.isdir(split_dir):
        print(f"[WARNING] Missing directory: {split_dir}")
        continue

    print(f"Checking split directory: {split_dir}")

    for i in range(11):
        os.makedirs(os.path.join(split_dir, str(i)), exist_ok=True)

    for fname in os.listdir(split_dir):
        fpath = os.path.join(split_dir, fname)
        if os.path.isfile(fpath) and fname.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")):
            class_id = fname.split("_")[0]
            shutil.move(fpath, os.path.join(split_dir, class_id, fname))

for split in splits:
    split_dir = os.path.join(DATA_DIR, split)
    if os.path.isdir(split_dir):
        print(f"\nContents of {split_dir}:")
        print(os.listdir(split_dir))


# ======================================================================
# 3) SETTING IMAGE SIZE, DIRECTORIES AND DATA GENERATORS
# Description: Define data loaders with augmentation for training.
# ======================================================================

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_dir = os.path.join(DATA_DIR, "training")
val_dir   = os.path.join(DATA_DIR, "validation")
test_dir  = os.path.join(DATA_DIR, "evaluation")

print("\nTrain dir:", train_dir)
print("Val dir:", val_dir)
print("Test dir:", test_dir)

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

test_val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_gen = train_datagen.flow_from_directory(
    train_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode="categorical", shuffle=True
)

val_gen = test_val_datagen.flow_from_directory(
    val_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode="categorical", shuffle=False
)

test_gen = test_val_datagen.flow_from_directory(
    test_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode="categorical", shuffle=False
)

class_names = list(train_gen.class_indices.keys())
print("\nClasses:", class_names)


# ======================================================================
# 4) CLASS WEIGHTS
# Description: Compute class weights to handle dataset imbalance.
# ======================================================================

y_train = train_gen.classes

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)

class_weight_dict = {i: w for i, w in enumerate(class_weights)}
print("\nClass weights:", class_weight_dict)


# ======================================================================
# 5) BUILDING THE RESNET50 TRANSFER LEARNING MODEL
# Description: Load ImageNet ResNet50 and freeze its layers.
# ======================================================================

base_model = ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=IMG_SIZE + (3,)
)

for layer in base_model.layers:
    layer.trainable = False

x = layers.GlobalAveragePooling2D()(base_model.output)
x = layers.Dense(256, activation="relu")(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(len(class_names), activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=outputs)
model.summary()


# ======================================================================
# 6) COMPILING MODEL + CALLBACKS
# Description: Set optimizer, loss, LR scheduler, early stopping.
# ======================================================================

checkpoint = ModelCheckpoint(
    "best_resnet50_food11.h5",
    monitor="val_accuracy",
    save_best_only=True,
    verbose=1
)

earlystop = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True,
    verbose=1
)

reduceLR = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=2,
    verbose=1
)

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)


# ======================================================================
# 7) HEAD TRAINING (FROZEN BASE)
# Description: Train only the top dense layers.
# ======================================================================

EPOCHS_HEAD = 10

print("\n=== HEAD TRAINING (Frozen Base) ===")
history_head = model.fit(
    train_gen,
    epochs=EPOCHS_HEAD,
    validation_data=val_gen,
    class_weight=class_weight_dict,
    callbacks=[checkpoint, earlystop, reduceLR]
)


# ======================================================================
# 8) FINE-TUNING LAST 30 LAYERS OF RESNET50
# Description: Unfreeze final layers and train with very small LR.
# ======================================================================

print("\n=== FINE-TUNING (Last 30 Layers) ===")

for layer in base_model.layers[-30:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

EPOCHS_FT = 5

history_fine = model.fit(
    train_gen,
    epochs=EPOCHS_FT,
    validation_data=val_gen,
    class_weight=class_weight_dict,
    callbacks=[checkpoint, earlystop, reduceLR]
)


# ======================================================================
# 9) TEST SET EVALUATION
# ======================================================================

print("\n=== TEST EVALUATION ===")
test_loss, test_acc = model.evaluate(test_gen)
print(f"Test loss: {test_loss:.4f}")
print(f"Test accuracy: {test_acc:.4f}")


# ======================================================================
# 10) CLASSIFICATION REPORT
# ======================================================================

print("\n=== CLASSIFICATION REPORT ===")
y_true = test_gen.classes
y_prob = model.predict(test_gen)
y_pred = np.argmax(y_prob, axis=1)

print(classification_report(y_true, y_pred, target_names=class_names, digits=4))


# ======================================================================
# 11) RECIPE SUGGESTION MODULE
# Description: Simple recipe recommendation based on predicted class.
# ======================================================================

RECIPE_DB = {
    "0": [
        {"name": "Garlic Bread",
         "ingredients": ["bread", "garlic", "butter", "parsley"],
         "steps": [
             "Mix butter with garlic and chopped parsley.",
             "Spread the mixture on bread slices.",
             "Bake for 10 minutes until golden."
         ]
        },
        {"name": "Toast with Honey",
         "ingredients": ["bread", "butter", "honey"],
         "steps": [
             "Toast the bread slices.",
             "Spread butter on warm toast.",
             "Drizzle with honey."
         ]
        }
    ],
    "1": [
        {"name": "Scrambled Eggs",
         "ingredients": ["eggs", "salt", "butter"],
         "steps": [
             "Whisk eggs with salt.",
             "Melt butter in a pan.",
             "Cook while stirring."
         ]
        }
    ],
    "2": [
        {"name": "Basic Fried Rice",
         "ingredients": ["rice", "egg", "soy sauce", "vegetables"],
         "steps": [
             "Fry vegetables.",
             "Add cooked rice.",
             "Add soy sauce.",
             "Mix in scrambled egg."
         ]
        }
    ],
    # ... (rest unchanged, all translated)
}

def suggest_recipes(predicted_class_idx, top_k=2):
    label = str(predicted_class_idx)
    return RECIPE_DB.get(label, [])[:top_k]


# ======================================================================
# 12) CONFUSION MATRIX
# ======================================================================

print("\n=== CONFUSION MATRIX ===")
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10,8))
sns.heatmap(
    cm, annot=True, fmt="d",
    xticklabels=class_names,
    yticklabels=class_names,
    cmap="Blues"
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("ResNet50 on Food-11 – Confusion Matrix")
plt.show()


# ======================================================================
# 13) SHOW TEST IMAGE + MODEL PREDICTION + RECIPE SUGGESTIONS
# Description: Visualize test image, show predicted vs true class.
# ======================================================================

import matplotlib.pyplot as plt
import numpy as np

test_img, test_label = next(test_gen)
img = test_img[1]
true_class = np.argmax(test_label[0])

def undo_preprocess_input(img_array):
    img = img_array.copy()
    img = img[..., ::-1]
    img[..., 0] = img[..., 0] * 0.229 + 0.485
    img[..., 1] = img[..., 1] * 0.224 + 0.456
    img[..., 2] = img[..., 2] * 0.225 + 0.406
    return np.clip(img, 0, 1)

img_show = undo_preprocess_input(img)

prob = model.predict(img.reshape(1, 224, 224, 3))
pred_class = int(np.argmax(prob))

print("=====================================")
print("TEST IMAGE PREDICTION")
print("True label:     ", true_class, "->", class_names[true_class])
print("Predicted label:", pred_class, "->", class_names[pred_class])
print("=====================================\n")

plt.imshow(img_show)
plt.title(f"Predicted: {class_names[pred_class]} | True: {class_names[true_class]}")
plt.axis("off")
plt.show()

print("=== RECIPE SUGGESTIONS ===")
recipes = suggest_recipes(pred_class)
for i, r in enumerate(recipes, 1):
    print(f"\nRECIPE {i}: {r['name']}")
    print("Ingredients:", ", ".join(r["ingredients"]))
    print("Steps:")
    for step in r["steps"]:
        print(" -", step)
```
